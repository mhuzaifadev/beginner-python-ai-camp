## üß† Why Functions?  
Imagine you have one big homework.  
Is it easier to do the entire thing at once, or do it in small chunks? ü§î  

**Exactly ‚Äî small chunks make things simpler!**  
That‚Äôs what a **function** is in Python.

## üéØ Functions with Parameters  
You can pass **information** into a function so it can do more!

For example:


```python
# üê£ Function with parameters
def say_hello(name):
    print(f"Hello, {name}! üëã")

say_hello("Ali")
say_hello("Sara")
```

## üîß Built-in Modules
Python already has **ready-made tools** you can use, called modules.

Some cool ones:
- `math`
- `random`
- `datetime`

(üëâ Check this link to read the full official Python module list):

```python
# üê£ math module
import math

print(math.sqrt(16))   # 4.0
print(math.pi)         # 3.14159...
print(math.factorial(5))  # 120
```

```python
# üê£ datetime module
from datetime import datetime

now = datetime.now()
print("Current date and time is:", now)
print("Year:", now.year)
print("Month:", now.month)
print("Day:", now.day)
```

## üîÑ Recursion ‚Äì A function that calls itself
Imagine you're looking for a word in a dictionary:
- If you haven't found it, look on the next page.
- Repeat until you find the word.

That‚Äôs kind of like **recursion**!

üí° Let's see a simple example ‚Äî finding the factorial of a number.

<img src = "https://www.qodo.ai/wp-content/uploads/2023/05/post-1496-1.jpg">

```python
# üê£ Recursion example: factorial
def factorial(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial(n-1)

print(factorial(5))  # 120
print(factorial(3))  # 6
```

## üß† Practice Time ‚Äî Functions & Recursion
‚úÖ **Practice Exercises:**

1. üßÆ Write a function that returns the sum of all numbers from 1 to `n`.
2. üîÑ Write a recursive function to print the numbers 1 to 10.
3. ‚úèÔ∏è Write a function to check if a word is a palindrome (e.g. "level").
4. üé≤ Write a function `roll_until_6()` that keeps rolling a dice using `random.randint(1, 6)` until it gets 6. Print all the rolls!

*(Hint: use loops and `my_toolbox.roll_dice()` if you want to use your own module!)*


```python
# üê£ Practice example 1: sum of 1 to n
def sum_to_n(n):
    return sum(range(1, n+1))

print(sum_to_n(10))  # 55
```

```python
# üê£ Practice example 2: recursion to print 1 to 10
def print_numbers(n=1):
    if n <= 10:
        print(n)
        print_numbers(n+1)

print_numbers()
```

```python
# üê£ Practice example 3: palindrome check
def is_palindrome(word):
    return word == word[::-1]

print(is_palindrome("level"))   # True
print(is_palindrome("apple"))   # False
```

```python
# üê£ Practice example 4: roll_until_6
import random

def roll_until_6():
    roll = random.randint(1, 6)
    print(f"Rolled a {roll}")
    if roll != 6:
        roll_until_6()

roll_until_6()
```
